```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure


%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #Adjusts the configurations of the plots


#Read in the data

df=pd.read_csv(r'C:\Users\palla\OneDrive\Desktop\Portfolio\movies.csv')
```

```python
df.head()
```

```python
# Finding missing values

for column in df.columns:
    missing_percentage = np.mean(df[column].isnull())
    print('{} - {}%'.format(column,round(missing_percentage,2)))
    
```

```python
df.dtypes
```

```python
df = df.dropna()
```

```python
# update release year based on released column value

df['updated_release_year']= df['released'].astype(str).str.partition(',')[2].str[:5]
df.head()
```

```python
# sort gross column
df=df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#Display all rows
pd.set_option('display.max_rows',None)
```

```python
#Finding if any duplicates
#df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scattered Plot bugdet vs gross

plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget Vs Gross Earnings')
plt.xlabel('Budget for a film')
plt.ylabel('Gross income for a film')
plt.show()

```

```python
df.head()
```

```python
#Regression Plot

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"red"},line_kws={"color":"blue"})
```

```python
#Find Correlation

#df.corr()
df.corr(method='pearson')#'kendall','spearman'
```

```python
#Correlation matrix
correlation_matrix=df.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for numeric values')
plt.show()
```

```python
#Create new dataframe where object types convereted into numeric for creating correlation matrix

df_numerize = df
for col_names in df_numerize.columns:
    if(df_numerize[col_names].dtype=='object'):
        df_numerize[col_names]=df_numerize[col_names].astype('category')
        df_numerize[col_names]=df_numerize[col_names].cat.codes
    
df_numerize.head()    
```

```python
df.head()
```

```python
#Correlation matrix
correlation_matrix=df_numerize.corr(method='pearson')
sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation Matrix for numeric values')
plt.show()
```

```python
#Sorting the correlation and find higher correlation with  value >0.5

correlation_mat =  df_numerize.corr()
corr_pairs=correlation_mat.unstack() #creates pairs
sorted_pair=corr_pairs.sort_values()

#High correlation
high_corr=sorted_pair[(sorted_pair)>0.5]
high_corr
```
